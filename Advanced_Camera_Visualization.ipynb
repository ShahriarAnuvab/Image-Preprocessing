```


# Importing all necessary libraries
import cv2
import os

# Read the video from specified path
cam = cv2.VideoCapture("gdrive/My Drive/Advanced Camera Dataset/videoplayback.mp4")

try:

    # creating a folder named data
    if not os.path.exists('gdrive/My Drive/Advanced Camera Dataset/data'):
        os.makedirs('gdrive/My Drive/Advanced Camera Dataset/data')

# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')

# frame
currentframe = 0

while(True):

    # reading from frame
    ret,frame = cam.read()

    if ret:
        # if video is still left continue creating images
        name = 'gdrive/My Drive/Advanced Camera Dataset/data/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name)

        # writing the extracted images
        cv2.imwrite(name, frame)

        # increasing counter so that it will
        # show how many frames are created
        currentframe += 1
    else:
        break

# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_root = "/content/gdrive/MyDrive/Advanced Camera Dataset/data"
class_yolo = "/content/yolov3.txt"
config_yolo = "/content/yolov3.cfg"
weights_yolo = "/content/yolov3.weights"

for img in os.listdir(img_root):
  image = cv2.imread(f"{img_root}/{img}")
  plt.imshow(image)
  break
```

```
def get_output_layers(net):

    layer_names = net.getLayerNames()
    try:
        output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
    except:
        output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]

    return output_layers


def draw_prediction(img, class_id, confidence, x, y, x_plus_w, y_plus_h):

    label = str(classes[class_id])

    color = COLORS[class_id]

    cv2.rectangle(img, (x,y), (x_plus_w,y_plus_h), color, 2)

    cv2.putText(img, label, (x-10,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)


for img in os.listdir(img_root):
  image = cv2.imread(f"{img_root}/{img}")

  Width = image.shape[1]
  Height = image.shape[0]
  scale = 0.00392

  classes = None

  with open(class_yolo, 'r') as f:
      classes = [line.strip() for line in f.readlines()]

  COLORS = np.random.uniform(0, 255, size=(len(classes), 3))

  net = cv2.dnn.readNet(weights_yolo, config_yolo)

  blob = cv2.dnn.blobFromImage(image, scale, (416,416), (0,0,0), True, crop=False)

  net.setInput(blob)

  outs = net.forward(get_output_layers(net))

  class_ids = []
  confidences = []
  boxes = []
  conf_threshold = 0.5
  nms_threshold = 0.4


  for out in outs:
      for detection in out:
          scores = detection[5:]
          class_id = np.argmax(scores)
          confidence = scores[class_id]
          if confidence > 0.5:
              center_x = int(detection[0] * Width)
              center_y = int(detection[1] * Height)
              w = int(detection[2] * Width)
              h = int(detection[3] * Height)
              x = center_x - w / 2
              y = center_y - h / 2
              class_ids.append(class_id)
              confidences.append(float(confidence))
              boxes.append([x, y, w, h])


  indices = cv2.dnn.NMSBoxes(boxes, confidences, conf_threshold, nms_threshold)

  for i in indices:
      try:
          box = boxes[i]
      except:
          i = i[0]
          box = boxes[i]

      x = box[0]
      y = box[1]
      w = box[2]
      h = box[3]
      draw_prediction(image, class_ids[i], confidences[i], round(x), round(y), round(x+w), round(y+h))


  cv2.imwrite(f"{img_root}/{img}_obj_det", image)
  print ('Creating...' + f"{img_root}/{img}_obj_det")

cv2.destroyAllWindows()
```
